```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=7f7e0135c769a1b12a7b3744a3a46cc4&language=en-US&page=1&region=AT')
```

```python
# Response types -
# 1. 200 - ok
# 2. 404 - Not available
# 3. 500 - server down
```

```python
response.json()['results']
```

```python
# Convert the request json file to a datframe
pd.DataFrame(response.json()['results'])
```

```python
# # We can choose specific columns which were only required
df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
# This is only for page 1 there are a total of 528 pages
# so we'll do this work in a loop
```

```python
df = pd.DataFrame()
```

```python
for i in range(1,529):
    # 1. Asking for response through api for page i
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=dd5ac7ff4913330e7884e31313a9b3d4&language=en-US&page={}&region=AT'.format(i))
    # 2. convert json data to datframe
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = pd.concat([df,temp_df], ignore_index=True)
    
```
