```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python

df = pd.read_csv('train (1).csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,cv=20,scoring='accuracy'))
```

## Applying Feature Construction

```python
X['family_size'] = X['SibSp']+X['Parch']+1
# Family = sibling&spouse + parent&child + himself
```

```python
X.head()
```

```python
def family(num):
    if num==0:
        # Alone
        return 0
    elif num>1 and num<=4:
        # Samll family
        return 1
    else:
        # Large family
        return 2
```

```python
family(4)
```

```python
X['Family_type'] = X['family_size'].apply(family)
```

```python
X.head()
```

```python

X.drop(columns=['SibSp','Parch','family_size'],inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,cv=20,scoring='accuracy'))
```

```python
# Here due to Feature construction ,accuracy rose from 69 to 73
# So everytime working on data , keep checking if i can apply feature construction or not
```
