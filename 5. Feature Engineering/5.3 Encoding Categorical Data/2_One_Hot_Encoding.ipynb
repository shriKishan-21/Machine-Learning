### 1. One Hot Encoding using pandas

```python
# Pnadas use get_dummies() function to perform One Hot Encoding
pd.get_dummies(df,columns=['fuel','owner'])
```

```python
# But here we did'nt solve multicollinearity issue
```

### 2. K-1 One Hot Encoding

```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
```

```python
# Form both the new columns created for each category, one column is redundant as it can be derived from the other column. 
# To avoid this, we can use drop_first=True parameter in get_dummies() function which will drop the first column of each category.
```

```python
# But we do not use pd.get_dummies() in Machine Learning because it does not remember which column were placed in which position .
# For each time we can get different result.
# so we use sklearn OneHotEncoder which remember the position of each column.
```

### 3. One Hot Encoding using sklearn

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df.drop('selling_price',axis=1),df['selling_price'],test_size=0.2,random_state=42)

x_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

# To drop the first column to avoid multicollinearity - pass the hyperparameter drop='first'
ohe = OneHotEncoder()
```

```python
# Here as we only need to apply ohe to specific columns.
# For that we have to remove those columns and transform them seperately and then concatenate them back to the dataframe.
# Here column transform is used which do all the fitting and transforming in one step. But we will use it later .
```

```python
x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
```

```python
# It craetes a sparse matrix to save memory. To convert it to array use .toarray() method
x_train_new.toarray().shape
```

```python
x_test_new = ohe.transform(x_test[['fuel','owner']])
```

```python
x_train[['brand','km_driven']].values
```

```python
# now both are numpy arrays so we can hstack them together
np.hstack((x_train[['brand','km_driven']].values , x_train_new.toarray())).shape
```

### Using hyperparameter

#### A. drop='first'

```python
# To drop the first column to avoid multicollinearity - pass the hyperparameter drop='first'
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first')
x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
```

```python
x_train_new.toarray().shape
```

```python
x_test_new = ohe.transform(x_test[['fuel','owner']])
x_train[['brand','km_driven']].values
```

```python
np.hstack((x_train[['brand','km_driven']].values , x_train_new.toarray())).shape
```

#### B. sparse_output = False

```python
# Using sparse_output = False we do not need to use .toarray() method
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',sparse_output=False)
x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
```

```python
x_train_new.shape
```

```python
x_test_new = ohe.transform(x_test[['fuel','owner']])
x_train[['brand','km_driven']].values
```

```python
np.hstack((x_train[['brand','km_driven']].values , x_train_new)).shape
```

#### C. dtype = np.int32

```python
# using dtype = np.int32 we can change the datatype to save memory
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
```

```python
x_train_new
```

```python
x_test_new = ohe.transform(x_test[['fuel','owner']])
x_train[['brand','km_driven']].values
```

```python
np.hstack((x_train[['brand','km_driven']].values , x_train_new))
# Here  we got int datatype
```
