### **Working with Dates**

```python
# Converting to datetime dtype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
date.sample()
```

#### 1. Extract year

```python
date['date_year'] = date['date'].dt.year
```

```python
date.sample()
```

#### 2. Extract month

```python
date['date_month'] = date['date'].dt.month
```

```python
date.sample()
```

#### 3. Extract day

```python
date['date_day'] = date['date'].dt.day
```

```python
date.sample()
```

#### 4. Days of week

```python
date['date_dow'] = date['date'].dt.dayofweek
date.sample()
```

#### 5. Days of week name

```python
date['date_down'] = date['date'].dt.day_name()
date.sample()
```

#### 6. is Weekend?

```python
date['date_is_weekend'] = np.where(date['date_down'].isin(['saturday','sunday']),1,0)
date.sample()
```

#### 7. Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week
date.sample()
```

#### 8. Extract quarter

```python
date['quarter'] = date['date'].dt.quarter
date.sample()
```

#### 9. Extract semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.sample()
```

## **Extratct time elapsed between dates**

```python
import datetime

today = datetime.datetime.today()

today
```

```python
# Date difference till today
date['diff'] = today - date['date']
date.sample()
```

```python
# only dates required in that
(today - date['date']).dt.days
```

```python
# Month passed
np.round((today - date['date'])/np.timedelta64(1,'m'),1)
```
