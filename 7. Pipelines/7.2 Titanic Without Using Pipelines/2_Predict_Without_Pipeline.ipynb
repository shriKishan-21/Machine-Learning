```python
import pickle
import numpy as np
```

```python
ohe_sex = pickle.load(open('models/ohe_sex.pkl', 'rb'))
ohe_embarked = pickle.load(open('models/ohe_embarked.pkl', 'rb'))
clf = pickle.load(open('models/clf.pkl', 'rb'))
```

```python
# Assume user input 
# Pclass/gender/age/sibsp/Fare/Embarked
test_input = np.array([2,'male',31.0,0,0,10.5,'S'],dtype=object).reshape(1,7)
```

```python
test_input
```

```python
# Here for testing the model , we have to follow the same steps with the input data as we did while training model
# 1. use imputer to handle missing value - no need here since no missing value input
# 2. ohe sex column
# 3. ohe embarked column
# Order should be same as model training - most important
```

```python
test_input_sex = ohe_sex.transform(test_input[:,1].reshape(1,1))
```

```python
test_input_sex
```

```python
test_input_embarked = ohe_embarked.transform(test_input[:,-1].reshape(1,1))
```

```python
test_input_embarked
```

```python
test_input_age = test_input[:,2].reshape(1,1)
```

```python
# Here the order of conactenation should be same as traing model
test_input_transformed = np.concatenate((test_input[:,[0,3,4,5]],test_input_age,test_input_sex,test_input_embarked),axis=1)
```

```python
test_input_transformed
```

```python
clf.predict(test_input_transformed)
```
