```python
from sklearn.datasets import make_classification
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x,y = make_classification(n_samples=5,n_classes=2,n_features=2,n_informative=2,n_redundant=0,random_state=0)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
clf.feature_importances_
```

```python
# For more samples
```

```python
x,y = make_classification(n_samples=15,n_classes=2,n_features=2,n_informative=2,n_redundant=0,random_state=0)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
clf.feature_importances_
```

```python
# Feature importance in Random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=2)
rf.fit(x,y)
```

```python
rf.feature_importances_
```

```python
# Importance of 0th feature
print(rf.estimators_[0].feature_importances_)
```

```python
# Importance of 1th feature
print(rf.estimators_[1].feature_importances_)
```

```python
# Feature importance of 0th feature in random forest - avg of both decision tree
print((0.42307692+0.81481481)/2)
```

```python
# Feature importance of 1th feature in random forest - avg of both decision tree
print((0.57692308+0.18518519)/2)
```
