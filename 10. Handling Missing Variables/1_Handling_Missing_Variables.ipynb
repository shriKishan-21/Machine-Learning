```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passenger travelling with')
```

```python
# For number column
```

```python
# Extract Numerical Part
df['numerical_part'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
# Extract Categorical Part
df['categorical_part'] = np.where(df['numerical_part'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
# For Cabin Column
```

```python
# Capture Numerical Values
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['cabin_cat'] = df['Cabin'].str[0]
```

```python
df.sample(5)
```

```python
# For Ticket column
```

```python
# Extract the Numerical Part
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')

# Extract the categorical part
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(),np.nan,df['ticket_cat'])
```
